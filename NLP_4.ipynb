```python
from urllib import request
```

```python
url = "https://www.gutenberg.org/files/98/98-0.txt"
```

```python
response = request.urlopen(url)
```

```python
raw = response.read().decode('utf8')
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
```

```python
print(tokens[:20])
```

```python
#BeautifulSoup
#preprocessing - RE to clean any html tags or unnecessary characters
#PoS tagging
```

```python
#Lancaster, porter, snowball,regex
```

```python
import nltk
from nltk.stem import PorterStemmer

porter = PorterStemmer()
porter.stem('cacti')
porter.stem('happiness')
porter.stem('joyous')

#happiness,joyous, cactii
```

```python
import nltk
from nltk.stem import LancasterStemmer

lanc = LancasterStemmer()
lanc.stem('singing')
```

```python
from nltk.stem import RegexpStemmer
regexp = RegexpStemmer('ing$')
print(regexp.stem('singing'))
print(regexp.stem('walking'))
print(regexp.stem('running'))
```

```python
from nltk.stem import SnowballStemmer

snow = SnowballStemmer('spanish')
snow.stem('estudiantes')
```

```python
for words in tokens:
    print(lanc.stem(words))
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
print(lemma.lemmatize("am",pos='v'))
print(lemma.lemmatize("am"))
print(lemma.lemmatize("cacti"))
print(lemma.lemmatize("mice"))
print(lemma.lemmatize("good"))
print(lemma.lemmatize("better",pos = 'a'))
print(lemma.lemmatize("best",pos = 'a'))
```
