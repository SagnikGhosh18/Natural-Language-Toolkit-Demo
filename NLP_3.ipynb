## Task 1

```python
import nltk
```

```python
def gender_features(word):
    return{'last_letter':word[-1]}
```

```python
gender_features('Winston')
```

```python
from nltk.corpus import names
```

```python
labeled_names=([(name,'male') for name in names.words("male.txt")] + [(name,'female') for name in names.words("female.txt")])
```

```python
import random
random.shuffle(labeled_names)
```

```python
print(labeled_names[0:10])
```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
```

```python
type(featuresets)
```

```python
print(featuresets)
```

```python
len(featuresets)
```

```python
train_set,test_set = featuresets[500:],featuresets[:500]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('David'))
```

```python
classifier.classify(gender_features('Sagnik'))
```

```python
classifier.classify(gender_features('Eshaan'))
```

```python
classifier.classify(gender_features('Somit'))
```

```python
classifier.classify(gender_features('Shobhit'))
```

```python
classifier.classify(gender_features('Ritu'))
```

```python
classifier.classify(gender_features('Sakshi'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
